### Calculer les distances des trajets

On veut maintenant le prix moyen et la durée moyenne selon le moyen de transport, par distance de trajet (0-200km, 201-800km, 800-2000km, 2000+km). On a deux manières de voir les choses : ou bien considérer un trajet entre une ville A et une ville B comme étant la distance entre ces deux villes, ou bien prendre en compte les escales. Commençons par mettre en place la première méthode. 

#### Distance directe

D'abord, on veut une fonction de distance, qui prend en argument deux villes, et qui retourne leur distance en km. Soient deux villes de latitude $l_1$, $l_2$, et de longitude $L_1$, $L_2$, et $R$ le rayon de la terre. Alors, la distance curviligne entre ces villes est $\arccos(\sin(l_1)\sin(l_2) + \cos(l_1)\cos(l_2)\cos(L_1-L_2))\times R$. Attention : les angles doient être exprimés en radians !

```python
def dist(a,b):
    if a == b: return 0
    l1 = np.radians(stations.loc[a]["latitude"])
    l2 = np.radians(stations.loc[b]["latitude"])
    L1 = np.radians(stations.loc[a]["longitude"])
    L2 = np.radians(stations.loc[b]["longitude"])
    distance = math.acos(math.sin(l1) * math.sin(l2) + \
                         math.cos(l1) * math.cos(l2) * math.cos(L1 - L2)) * 6371
    return distance    
```

```python
ticket_data["journey_distance"] = list(map(dist, ticket_data["o_station"], ticket_data["d_station"] ))
ticket_data
```

Regardons quelques exemples de distances :

```python
for i in range (10): 
    print("La distance de " + \
          str(stations.loc[ticket_data["o_station"].iloc[50*i]]["unique_name"]) + \
          " à " + \
          str(stations.loc[ticket_data["d_station"].iloc[50*i]]["unique_name"]) + \
          " est de " + \
          str(dist(ticket_data["o_station"].iloc[50*i], ticket_data["d_station"].iloc[50*i])) + \
          "km. \n")
```

#### Distance en prenant en compte les escales

Maintenant, on va chercher à calculer la distance en prenant en compte les escales. Pour cela, on a besoin de se concentrer sur une ligne du dataframe ``ticket_data``, et non plus sur les villes de départ et d'arrivée. On remarque que les ``middle_stations`` sont des chaînes de caractères, et non des listes d'entiers. On change donc ça :

```python
def str2int(data):
    liste = []
    integer = ''
    for x in data[1:]:
        if (x == ',') or (x =='}'):
            liste.append(int(integer))
            integer = ''
        else:
            integer+=x
    return liste
```

La fonction ``str2int`` transforme une chaîne de caractères de type '{1,23}' en la liste d'entiers \[1,23\].

```python
ticket_data["middle_stations_list"] = list(map(str2int, ticket_data["middle_stations"]))
ticket_data
```

On a bien le changement désiré. Maintenant, il s'agit de calculer pour chaque trajet la distance parcourue en prenant en compte les escales.

```python
def dist_with_stopover(departure, stopovers, arrival):
    journey = [departure] + stopovers + [arrival]
    distance = 0
    for i in range(len(journey) - 1):
        distance += dist(journey[i], journey[i+1])
    return distance
```

```python
ticket_data["journey_distance_stopover"] = \
    list(map(dist_with_stopover, ticket_data["o_station"], 
             ticket_data["middle_stations_list"], 
             ticket_data["d_station"]))

ticket_data
```

On remarque que les escales ont plus ou moins d'effet sur la distance totale :
1. Aller directement de Aix-en-Provence (PACA, France) à Saint-Martin-sur-le-Pré (Marne, France) ne demande que 4km de moins qu'avec un détour par les Alpes (Chamonix et Bourg-Saint-Maurice) (trajet de la première ligne).
2. Par contre, aller directement de Saint-Priest-Taurion (Haute-Vienne, France) à Montpezat de Quercy (Tarn-et-Garonne) est plus de deux fois plus court (504km contre 1085km) que de passer par le Mont-Saint-Michel (Manche, France), puis Paris (Paris, France), puis Marnoz (Jura, France) et enfin arriver à destination (trajet de la 2e ligne). 

### Déterminer les moyens de transport

On va maintenant chercher à déterminer, pour chaque trajet, le moyen de transport principal (donné par ``company`` de ``ticket_data``) et les moyens de transport secondaires (donnés par ``other_companies``).

Comme pour les escales, on a besoin de changer le type de données des companies de transport utilisées dans ``ticket_data`` : de chaîne de caractères à liste d'entiers, comme précédemment pour les escales.

```python
ticket_data["other_companies_list"] = list(map(str2int, ticket_data["other_companies"]))
ticket_data
```

Maintenant, on implémente une fonction qui cherche le type de transport principal, ainsi que les potentiels secondaires.

```python
def transport_mean(company, others):
    main = [providers["transport_type"].loc[company]]
    others = [providers["transport_type"].loc[x] for x in others]
    return tuple(sorted(list(set(main + others))))#on utilise ici un tuple parce que python ne peut pas 
                                                  #groupby des objets unhashable, or les listes le sont
```

```python
ticket_data['transport_type'] = list(map(transport_mean, ticket_data["company"], 
                                         ticket_data["other_companies_list"]))
ticket_data
```

### Faire des catégories par distance

On veut connaître les temps de trajet et coûts moyens en fonction du type de transport et de la catégorie de la distance (moins de 200km, entre 201km et 800km, entre 801 et 2000km, au delà de 2001km). La question est : parle-t-on de la distance entre le point de départ et le point d'arrivée, ou de la distance totale du trajet (en comptant les escales donc) ? Pour moi, les deux peuvent être valables : 

1. La distance entre le point de départ et le point d'arrivée est finalement ce qui nous intéresse : notre désir initial était ce trajet là, et le surplus de distance peut être vu comme un désavantage
2. La distance en comprenant les escales est la distance effective, et il peut être intéressant de voir si ces distances sont moins onéreuses et longues à parcourir.

J'ai donc décidé d'implémenter les deux. Remarque : même en comptant les escales, il n'y a pas de trajet supérieur à 2000km. J'ai donc décidé de faire de nouvelles catégories de distances : moins de 150km, entre 150km et 500km, entre 500km et 1000km, au delà de 1000km. 

Voilà ma première implémentation, en prenant en compte les escales. La durée moyenne :

```python
duration_less_150km_stopovers = \
    ticket_data.query('journey_distance_stopover < 151') \
               .groupby(['transport_type'])['journey_in_sec'] \
               .mean()/3600

duration_151_500km_stopovers = \
    ticket_data.query('journey_distance_stopover >= 151 and journey_distance_stopover < 501') \
               .groupby(['transport_type'])['journey_in_sec'] \
               .mean()/3600

duration_501_1000km_stopovers = \
    ticket_data.query('journey_distance_stopover >= 501 and journey_distance_stopover < 1001') \
               .groupby(['transport_type'])['journey_in_sec'] \
               .mean()/3600.0
        
duration_more_1001km_stopovers = \
    ticket_data.query('journey_distance_stopover > 1001') \
               .groupby(['transport_type'])['journey_in_sec'] \
               .mean()/3600
```

Le prix moyen :

```python
price_less_150km_stopovers = \
    ticket_data.query('journey_distance_stopover < 151') \
               .groupby(['transport_type'])['price_in_cents'] \
               .mean()/100

price_151_500km_stopovers = \
    ticket_data.query('journey_distance_stopover >= 151 and journey_distance_stopover < 501') \
               .groupby(['transport_type'])['price_in_cents'] \
               .mean()/100

price_501_1000km_stopovers = \
    ticket_data.query('journey_distance_stopover >= 501 and journey_distance_stopover < 1001') \
               .groupby(['transport_type'])['price_in_cents'] \
               .mean()/100

price_more_1001km_stopovers = \
    ticket_data.query('journey_distance_stopover > 1001') \
               .groupby(['transport_type'])['price_in_cents'] \
               .mean()/100
```

Aussi, je pense qu'il serait intéressant de se demander ce qu'il se passe pour un trajet d'un point A à un point B distants de moins de 150km (par exemple), lorsqu'avec escale il fait moins de 150km, entre 150km et 500km, entre 500km et 1000km, ou encore au-delà de 1000km ?

On va implémenter cela, ainsi que l'effet d'un trajet d'un point A à un point B distants de moins 500km, lorsqu'avec escale il fait moins de 500km, entre 500km et 800km, entre 800km et 1200, ou au-delà de 1200km.

Ici, pour moins de 150km :

```python
duration_150km_less150km = \
    ticket_data.query('journey_distance < 151') \
               .query('journey_distance_stopover < 151') \
               .groupby(['transport_type'])['journey_in_sec'] \
               .mean()/3600
duration_150km_less500km = \
    ticket_data.query('journey_distance < 151') \
               .query('journey_distance_stopover >= 151 and journey_distance_stopover < 501') \
               .groupby(['transport_type'])['journey_in_sec'] \
               .mean()/3600
duration_150km_less1000km = \
    ticket_data.query('journey_distance < 151') \
               .query('journey_distance_stopover >= 501 and journey_distance_stopover < 1001') \
               .groupby(['transport_type'])['journey_in_sec'] \
               .mean()/3600
duration_150km_more1000km = \
    ticket_data.query('journey_distance < 151') \
               .query('journey_distance_stopover > 1001') \
               .groupby(['transport_type'])['journey_in_sec'] \
               .mean()/3600

price_150km_less150km = \
    ticket_data.query('journey_distance < 151') \
               .query('journey_distance_stopover < 151') \
               .groupby(['transport_type'])['price_in_cents'] \
               .mean()/100
price_150km_less500km = \
    ticket_data.query('journey_distance < 151') \
               .query('journey_distance_stopover >= 151 and journey_distance_stopover < 501') \
               .groupby(['transport_type'])['price_in_cents'] \
               .mean()/100
price_150km_less1000km = \
    ticket_data.query('journey_distance < 151') \
               .query('journey_distance_stopover >= 501 and journey_distance_stopover < 1001') \
               .groupby(['transport_type'])['price_in_cents'] \
               .mean()/100
price_150km_more1000km = \
    ticket_data.query('journey_distance < 151') \
               .query('journey_distance_stopover > 1001') \
               .groupby(['transport_type'])['price_in_cents'] \
               .mean()/100
```

# Conclusion

On commence à déceler un schéma (qui reste à confirmer) : pour des courts trajets, le bus ou le train semblent assez similaires, tant en prix qu'en durée.

Plus le trajet est long, plus le temps est réduit mais le prix augmenté en train. 

Faire des escales ne semble pas réellement valoir le coût : plus la distance d'un trajet est similaire à la distance à vol d'oiseau reliant l'origine à la destination, plus le prix est faible, et le temps court.

Cependant, d'autres critères seraient à évaluer, comme le moment de la réservation (plus on réserve tôt, plus les prix sont généralement faibles ; est-ce tout le temps vrai, pour tout mode de transport ?), le moment de l'année, la région, le nombre d'escales, le fait de changer de pays,...
